## Handling Imbalanced Dataset

1.) UP Sampling

2.) DOWN Sampling

```python
import numpy as np
import pandas as pd
from sklearn.utils import resample
```

```python
np.random.seed(23)

## Create a dataframe with two class
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)       ## 900
n_class_1 = n_samples - n_class_0                ## 100
```

```python
n_class_0, n_class_1
```

```python
## Creating my DataFrame with Imbalanced Dataset
## loc = 0 means it has a mean value 0 and
## scale = 1 means standard devaition 1
class_0 = pd.DataFrame({
    "Feature_1" : np.random.normal(loc = 0, scale = 1, size = n_class_0),
    "Feature_2" : np.random.normal(loc = 0, scale = 1, size = n_class_0),
    "Target" : [0] * n_class_0
})

class_1 = pd.DataFrame({
    "Feature_1" : np.random.normal(loc = 0, scale = 1, size = n_class_1),
    "Feature_2" : np.random.normal(loc = 0, scale = 1, size = n_class_1),
    "Target" : [1] * n_class_1
})
```

```python
df = pd.concat([class_0, class_1]).reset_index(drop = True)
df.head()
```

```python
df.tail()
```

```python
df['Target'].value_counts()
```

### Up Sampling

```python
df_minority = df[df['Target'] == 1]
df_majority = df[df['Target'] == 0]
```

```python
df_minority_upsampled = resample(df_minority, replace=True,    ## Sample with replacement
         n_samples= len(df_majority),
         random_state= 45)      
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.head()
```

```python
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
```

```python
df_upsampled['Target'].value_counts()
```

## Down Sampling

```python
np.random.seed(123)

n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0

class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'Target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'Target': [1] * n_class_1
})

df = pd.concat([class_0, class_1]).reset_index(drop=True)

print(df['Target'].value_counts())
```

```python
df_minority = df[df['Target'] == 1]
df_majority = df[df['Target'] == 0]
```

```python
df_majority_downsampled = resample(df_majority, replace=False,    ## Sample with replacement
         n_samples= len(df_minority),
         random_state= 45)      
```

```python
df_majority_downsampled.shape
```

```python
df_downsampled = pd.concat([df_minority, df_majority_downsampled])
```

```python
df_downsampled.Target.value_counts()
```
