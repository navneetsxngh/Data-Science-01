```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
## Encoding ---> Transform your Categorical data into Numerical data.
# (1). Label ---> It will encode data 1-1 columns
# (2). Ordinal ---> It will encode data more than 1 columns.
# (3). OneHot ---> It will create new column based on sub-categories bit it has only values not column name.
# (4). get_dummies ---> It will create new column based on sub-categories bit it shows column name with values.
```

## **LabelEncoder**

```python
df = pd.read_csv("/content/toyota.csv")
df.head(3)
```

```python
lb = LabelEncoder()
```

```python
x = df.drop(columns = ['price'])
y = df['price']
```

```python
df['model'].value_counts()
```

```python
x['model'] = lb.fit_transform(x['model'])
x['transmission'] = lb.fit_transform(x['transmission'])
x['fuelType'] = lb.fit_transform(x['fuelType'])
```

```python
x.head(3)
```

```python
x['model'].value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 59)
```

```python
np.round(x_train.describe(), 2)
```

```python
sc = StandardScaler()
```

```python
x_train_sc = sc.fit_transform(x_train)
```

```python
x_train_new = pd.DataFrame(x_train_sc, columns = x.columns)
```

```python
np.round(x_train_new.describe(), 2)
```
