```python
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

### Insurance Dataset

```python
df = pd.read_csv(r"C:\Users\singh\OneDrive\Desktop\Python\Data\insurance - insurance.csv")
df.head()
```

```python
## Checking Missing Values
df.isnull().sum()
```

```python
df['region'].value_counts()
```

```python
ode_sex = OrdinalEncoder(categories=[['male', 'female']])
ode_smoker = OrdinalEncoder(categories=[['no', 'yes']])
ode_region = OrdinalEncoder(categories=[['southeast', 'southwest', 'northwest', 'northeast']])

df['sex'] = ode_sex.fit_transform(df[['sex']])
df['smoker'] = ode_smoker.fit_transform(df[['smoker']])
df['region'] = ode_region.fit_transform(df[['region']])
```

```python
x = df.drop(columns= ['charges'])
y = df['charges']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=56)
```

```python
from sklearn.compose import ColumnTransformer

transfomer = ColumnTransformer(transformers= 
                               [('tnf1', OrdinalEncoder(categories= [['male', 'female']]), ['sex']),
                                ('tnf2', OrdinalEncoder(categories=[['no', 'yes']]), ['smoker']),
                                ('tnf3', OrdinalEncoder(categories=[['southeast', 'southwest', 'northwest', 'northeast']]), ['region'])], 
                                remainder='passthrough')

new_df = transfomer.fit_transform(df)
features = transfomer.get_feature_names_out()

df = pd.DataFrame(new_df, columns=features)
df.head()
```
