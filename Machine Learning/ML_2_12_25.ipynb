```python
## Normalization ---> Min = 0 and Max = 1
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
mn = MinMaxScaler()
```

## Importing Churn Modeling Data

```python
df = pd.read_csv("/content/Churn_Modelling - Churn_Modelling.csv")
df
```

```python
df = df.drop(columns = ['Surname', 'CustomerId', 'Geography'])
df
```

```python
df['Gender'] = df['Gender'].map({"Male" : 1, "Female" : 0})
df
```

```python
x = df.drop(columns= ['Exited'])
y = df['Exited']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,
                                                    y,
                                                    test_size = 0.2,
                                                    random_state = 59)
```

```python
x_train_mn = mn.fit_transform(x_train)
```

```python
x_train_new = pd.DataFrame(x_train_mn, columns = x_train.columns)
```

```python
np.round(x_train_new.describe(), 2)
```

## Importing Tips Data using Seaborn

```python
df = sns.load_dataset('tips')
df
```

```python
df = df.drop(columns = ['day'])
```

```python
df['time'] = df['time'].map({"Dinner" : 0, "Lunch" : 1})
df['smoker'] = df['smoker'].map({"Yes" : 1, "No": 0})
df['sex'] = df['sex'].map({"Male" : 1, "Female" : 0})
```

```python
df.head(5)
```

```python
x = df.drop(columns = ['total_bill'])
y = df['total_bill']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,
                                                    y,
                                                    test_size = 0.2,
                                                    random_state = 80)
```

```python
x_train_mn = mn.fit_transform(x_train)
x_train_new = pd.DataFrame(x_train_mn, columns = x_train.columns)
```

```python
np.round(x_train_new.describe(), 2)
```
