```python
# Normal Distribution
# from central Tendency (mean, median, mode) ---> left and right side equal data points
# Identification --> mean = 0, standard Deviation  = 1
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/covid_toy.csv")
df.head(3)
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df = df.drop(columns = ['city', 'cough'])
df.head(2)
```

```python
df['gender'] = df['gender'].map({'Male':1, 'Female':0})
df['has_covid'] = df['has_covid'].map({'Yes':1, 'No':0})
df.head(2)
```

```python
x = df.drop(columns = ['has_covid'])
y = df['has_covid']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,
                                                    y,
                                                    test_size = 0.2,
                                                    random_state = 42)
```

```python
np.round(x_train.describe(), 2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train_sc = sc.fit_transform(x_train)    ## fit means learn the parameters and transform means apply on that data.
```

```python
x_train_new = pd.DataFrame(x_train_sc, columns = x_train.columns)
```

```python
x_train_new.head(3)
```

```python
np.round(x_train_new.describe(), 2)
```
