## Python Logging

Logging is a crucial aspect of any application, providing a way to track events, errors and operational information. Python's built-in logging module offers a flexible framework for emitting log messages from Python programs.

```python
import logging

## Configure the basic logging settings
logging.basicConfig(level= logging.DEBUG)

## Log Messages with different severity
logging.debug("This is a Debug Message")
logging.info("This is an info Message")
logging.warning("This is a Warning Message")
logging.error("This is an error message")
logging.critical("This is a Critical Message")
```

## Logging Levels (very important)

From least to most severe:

| Level      | When to use                         |
| ---------- | ----------------------------------- |
| `DEBUG`    | Detailed dev info (variables, flow) |
| `INFO`     | General app progress                |
| `WARNING`  | Something odd, but not broken       |
| `ERROR`    | Something failed                    |
| `CRITICAL` | App may crash                       |


```python
import logging

logging.basicConfig(
    level=logging.DEBUG,
    filename = 'app.log',
    filemode= 'w',
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    force=True
)

logging.debug("This is a Debug Message")
logging.info("This is an info Message")
logging.warning("This is a Warning Message")
logging.error("This is an error message")
logging.critical("This is a Critical Message")
```
