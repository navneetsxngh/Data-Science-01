```python
## Importing Libraries
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVR
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, r2_score

print("Libraries Successfully Imported")
```

```python
df = pd.read_csv(r"covid_toy.csv")
df.head()
```

```python
df['fever'] = df['fever'].fillna(df['fever'].mean())
```

```python
df.info()
```

```python
cat_columns = [column for column in df.columns if df[column].dtype == 'str']
lb = LabelEncoder()

for col in cat_columns:
    df[col] = lb.fit_transform(df[col])
```

```python
x = df.drop(columns= ['has_covid'])
y = df['has_covid']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
## Linear Kernel
linear = SVC(kernel='linear')
linear.fit(x_train, y_train)
liner_pred = linear.predict(x_test)
linear_accuracy = accuracy_score(y_test, liner_pred)
print("Linear Accuracy : ", linear_accuracy)

## Polynomial Kernal
poly = SVC(kernel='poly', degree=3)
poly.fit(x_train, y_train)
poly_pred = poly.predict(x_test)
poly_accuracy = accuracy_score(y_test, poly_pred)
print("Polynomial Accuracy : ", poly_accuracy)
```
